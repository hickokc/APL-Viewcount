```python
from bs4 import BeautifulSoup as bs
import urllib2
import requests
import re
import datetime as dt
import csv
```

```python
x = raw_input("Enter keyword: ")
l = str(x)
ur = 'https://www.youtube.com/results?search_query='
url = ur+l
r  = requests.get(url)
data = r.text

links = []
titles = []

soup = bs(data, 'lxml')
for element in soup.find_all('h3'):#,{'id':'video-title'}):
    href = (element.get('href'))
    title = (element.get('aria-label'))
    links.append(href)
    titles.append(title)
```
